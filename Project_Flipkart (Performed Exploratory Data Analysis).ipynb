# Finding the top 10 costly phones

```python
var1=df.groupby(["Product_Name","selling_price"]).sum().reset_index().sort_values(['selling_price'],ascending=False,ignore_index=True)
var2=var1.head(10)
var2.head(10)
```

```python
p1=var2[["Product_Name","selling_price"]]
p1
```

#I can see that selling price is maximum for this phone:APPLE iPhone 14 Pro (Deep Purple, 1 TB)	

```python
p1.plot.bar("Product_Name","selling_price",color=(['r','c']),width=0.3)
font1={"color":"red"}
plt.ylabel('Selling_price',fontsize=16,fontdict=font1)
plt.xlabel('Product',fontsize=14,fontdict=font1)
plt.grid()
plt.show()
```

# Finding the top 5 phones which have great battery power

```python
df[["Product_Name","battery_in_mAh"]].reset_index().sort_values(["battery_in_mAh"],ascending=False,ignore_index=True).head()
```

# Finding the top 10 phones which have battery size is < 4000 and > 1000

```python
battery =df[(df['battery_in_mAh']<4000) & (df['battery_in_mAh']>1000)]
result1=battery[["Product_Name","battery_in_mAh"]].head(10)
result1
#phones_with_greatbattery=result1.head(10)
```

```python
result1.plot.bar("Product_Name","battery_in_mAh",color=(['b','c']),width=0.3)
font1={"color":"red"}
plt.ylabel('battery_in_mAh',fontsize=16,fontdict=font1)
plt.xlabel('Product',fontsize=14,fontdict=font1)
plt.grid()
plt.show()
```

# Mobile phones which have maximum discount

```python
var5=df.groupby(["Product_Name",'discount']).sum().reset_index().sort_values(['discount'],ascending=False,ignore_index=True)
var5[["Product_Name",'discount']].head(10)
```

# Finding the top 10 phone which have selling price range lies between 10000 to 30000


```python
selling =df[(df['selling_price']<30000) & (df['selling_price']>10000)]
result1=selling[["Product_Name",'selling_price']]
result1.head(10)
```

# Finding the phones which have warranty > 12 months and rating > 4

```python
best_product1=df[(df['warranty_in_months']>12) & (df['rating']>4.5)]
best_product=best_product1[["Product_Name",'warranty_in_months']]
best_product
```

```python
best_product.plot.bar("Product_Name","warranty_in_months",color=(['y','c']),width=0.3)
font1={"color":"red"}
plt.ylabel('warranty_in_months',fontsize=16,fontdict=font1)
plt.xlabel('Product',fontsize=14,fontdict=font1)
plt.grid()
plt.show()
```

# Phones which have received maximum ratings

```python
best_phone_by_rating=df[["Product_Name","rating_count"]].value_counts().reset_index().sort_values(["rating_count"],ascending=False,ignore_index=True)
best_phone_by_rating
```

```python
bestie=best_phone_by_rating[["Product_Name","rating_count"]]
b1=bestie.head(10)
b1
```

```python
b1.plot.bar("Product_Name","rating_count",color=(['c','c']),width=0.3)
font1={"color":"black"}
font2={"color":"blue"}
plt.ylabel('Charges in Lakhs',fontsize=16,fontdict=font1)
plt.xlabel('region',fontsize=14,fontdict=font1)
plt.grid()
plt.show()
```
